Для удобной работы с датой и временем в Python есть модуль datetime. Данный модуль используется для работы со временем и датами, позволяя представлять данную информацию в наиболее удобной форме.


# DATE

Класс date используется  для представления данных о дате и включает информацию о годе, месяце и числе.

```python
from datetime import date
my_date1 = date(1992, 10, 6)    # При создании новой даты нужно указать год, месяц и день.
my_date2 = date(day=6, month=10, year=1992) # Можно использовать именные аргументы, но все три атрибута конструктора одинаковы

#my_date3 = date(1992, 14, 40) # В случае использования неверного значения возникнет ошибка (исключение) ValueError


print(my_date1)
print(my_date2)

print(type(my_date1)) # <class 'datetime.date'>
```

Пусть **birth** - экземпляр класса date, т.е. определенная дата \

Экземпляр класса date имеет следующие атрибуты:
*  **birth.year** — год даты
*  **birth.month** — месяц даты
*  **birth.day** — день даты

Экземпляр класса date имеет следующие методы: \
*  **birth.weekday()** - метод определяет день недели даты birth(возвращает индекс дня недели, индексация с 0)
*  **birth.isoweekday()** - метод определяет день недели даты obj (возвращает индекс дня недели, индексация с 1)
*  **birth.toordinal()** - метод возвращает григорианский порядковый номер даты, где 1 января 1 года имеет порядковый номер 1.
*  **date.fromordinal(number)** - обратная конвертация; метод класса вернет дату по ее номеру number
*  **birth.isoformat()** или **str(birth)** или **birth.__str__()** - методы эквивалентны, возвращают строку, представляющую дату в формате ISO 'YYYY-MM-DD'
*  **date.fromisoformat(string)** - обратное преобразование; если строка string содержит запись даты в iSO формате, метод вернет экземпляр класса date
* **birth.strftime(format)** - форматирование даты в строку по заданному формату(подробнее в разделе datatime)

\


*  **date.today()** - метод класса, возвращающий текущую дату
*  **date.min** - минимально возможная дата 0001-01-01
*  **date.max** - максимально возможная дата 9999-12-31

Объекты типа date и time можно сравнивать с помощью операторов **==, !=, <, >, <= и  >=** или использовать встроенные функции **min(), max(), sorted(), filter()** и т.д. \

Оба типа данных date и time являются неизменяемыми.

```python
a = date.fromisoformat('2024-06-01')
print(a)
```

```python
from datetime import date

weekdays = ['Понедельник', 'Вторник', 'Среда', 'Четверг', 'Пятница', 'Суббота', 'Воскресенье']

my_date = date.today() #сегодняшняя дата

print('Год =', my_date.year, 'Месяц =', my_date.month, 'День =', my_date.day)

print('День недели: ' + weekdays[my_date.weekday()])
```

```python
from datetime import date

#функция, возвращающая количество суббот между двумя датами data1, data2
def saturdays_between_two_dates(data1, data2):
    start, end = min(data1, data2), max(data1, data2) #тк даты могут передаваться в любом порядке, находим левую и правую границы временного диапазона
    alldays = [date.fromordinal(i) for i in range(start.toordinal(), end.toordinal() + 1)] #массив всех дат в диапазоне [start, end]
    saturdays = list(filter(lambda day: day.weekday() == 5, alldays)) #массив дат, день недели у которых суббота
    return len(saturdays) #возвращаем количество суббот

date1 = date(2021, 11, 1)
date2 = date(2021, 11, 22)

print(saturdays_between_two_dates(date1, date2))
```

# TIME

Тип данных (класс) time используется для представления данных о времени и включает информацию о часах, минутах, секундах и микросекундах.

Микросекунда (мкс) — единица времени, равная одной миллионной доле секунды ( 10 ^ -6 c)

Конструктор типа time сначала принимает часы, затем минуты, секунды, а уже потом микросекунды. Мы можем использовать именованные аргументы (hour, minute, second, microsecond), нарушая указанный порядок.

Указывая аргументы hour, minute, second, microsecond, не следует забывать про ограничения(hour < 24 и тд).

В отличие от дат (тип данных date), чтобы создать объект типа time, необязательно указывать все его атрибуты в конструкторе. Недостающие данные о времени автоматически заполняются нулями.


```python
from datetime import time

time1 = time(11, 20, 54, 1234)
time2 = time(12)
print(type(time1))
print(time1)
print(time2)
```

Пусть **lunch** - экземпляр класса time, т.е. определенная дата

Экземпляр класса time имеет следующие атрибуты:
*  **lunch.hour** — часы времени
*  **lunch.minute** — минуты времени
*  **lunch.second** — секунды времени
*  **lunch.microsecond** — микросекунды времени

Класс поддерживает методы:
*  **time.fromisoformat(string)** - принцип работы аналогичен с классом date
*  **lunch.strftime(format)** - принцип работы аналогичен с классом date

# DATETIME

Типы данных date и time позволяют работать по отдельности с датами и временами. Однако на практике чаще требуется работать одновременно и с датой, и со временем. Для таких целей используется тип данных datetime.

При создании новой даты-времени (тип datetime) нужно указать год, месяц, день, часы, минуты, секунды и микросекунды. При этом год, месяц и день являются обязательными, а часы, минуты, секунды и микросекунды необязательными.

```python
from datetime import datetime

my_datetime = datetime(1992, 10, 6, 9, 40, 23, 51204)    # создаем полную дату-время
only_date = datetime(2021, 12, 31)                       # создаем дату-время с нулевой временной информацией

print(my_datetime)
print(only_date)
print(type(my_datetime))
```

!! **Класс наследует все атрибуты и методы от классов date и time.**

Пусть **registration** - экземляр класса datetime

* **datetime.strptime(text, pattern)** - string parse time; метод позволяет извлечь из строки text информацию об объекте datetime, записанную в формате format

Метод strftime наследуется от классов date и time. А метод strptime есть только в этом классе

# TIMEDELTA

Тип данных **timedelta** представляет из себя временной интервал (разница между двумя объектами datetime или date) и используется для удобного выполнения различных манипуляций над типами datetime или date.

При создании объекта timedelta можно указать следующие аргументы:

* недели (weeks)
* дни (days)
* часы (hours)
* минуты (minutes)
* секунды (seconds)
* микросекунды (microseconds)
* миллисекунды (milliseconds)

Мы можем выбрать любые их сочетания для задания временного интервала, при этом все аргументы являются необязательными и по умолчанию равны 0.



```python
from datetime import timedelta

delta = timedelta(days=7, hours=20, minutes=7, seconds=17)

print(delta)
print(type(delta))
```

Тип timedelta внутренне хранит только сочетание **days, seconds, microseconds**.

 А остальные единицы конвертируются:
* weeks -> 7 days
* hours -> 3600 seconds
* minutes -> в 60 seconds


```python
#любые аргументы, кроме days, seconds, microseconds, объединяются и нормализуются в три результирующих сочетания.
from datetime import timedelta
delta1 = timedelta(weeks=1, hours=23, minutes=61)
delta2 = timedelta(hours=25)
delta3 = timedelta(minutes=300)

print(delta1, delta2, delta3, sep = '\n')
```

Также временной интервал (timedelta) может быть отрицательным.

```python
from datetime import timedelta
delta1 = timedelta(minutes=-40)
delta2 = timedelta(seconds=-10, weeks=-2)
print(delta1)
print(delta2)
```

Элемент класса timedelta имеет только эти атрибуты:
* **days**
* **seconds**
* **microseconds**

Остальные можно конвертировать.

```python
from datetime import timedelta

#функция подсчитывает количество часов, минут
def hours_minutes(td):
    return td.seconds // 3600, (td.seconds // 60) % 60

delta1 = timedelta(hours=25, minutes=30)
print(delta1)
print(delta1.days) #так можно
#print(delta1.minutes) #так нельзя
print(hours_minutes(delta1)) #Достать часы, минуты можно так

```

**obj.total_seconds()** - метод возвращает общее количество секунд, содержащееся во временном интервале obj

Временные интервалы (тип timedelta) можно сравнивать**(==, !=, <, >, <=, >=)**, как и любые другие типы данных.

**Операции над временными интервалами timedelta:**

*   Сложение и вычитание времененных интервалов друг с другом
*   Сложение и вычитание времененных интервалов с объектами datetime или date
*   Сложение и вычитание объектов datetime, date -> результат deltatime
*   Умножение временнного интервала на число
*   Деление(/ или //) временнного интервала на число
*   Деление(/, // или %) временнного интервала на временный интервал



```python
from datetime import date, timedelta, datetime

# Пример: генератор, который возвращает даты в диапазоне [start, end) с шагом step
def date_range(start, end, step=timedelta(days=1)):
    current = start
    while current < end:
        yield current
        current += step

start = datetime(2025, 5, 26)
end = datetime(2025, 5, 30)

for i in date_range(start, end):
    print(i)
print()
for i in date_range(start, end, timedelta(hours=12)):
    print(i)
```
