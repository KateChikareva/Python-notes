# Индексы, срезы и маска

Обращение к элем по индексу и взятие среза работает также как в обычном питоне.

```python
a = np.arange(6)
print(a)
print(a[0:4])
print(a[0])

a[0:4] = 0  # Неочевидная вещь у слайсов np массива
print(a)

```

```python
# еще одна неочевидная особенность np массивов
a = np.array([1, 2, 3, 4, 5])
b = a[0:3]
b[0] = 10  # Если сделать слайс и положить его в другую переменную, обе переменные ссылаются на один и тот же массив в памяти
a[1] = 20  # поэтому эти изменения работают для двух массивов
print(a, b)

#чтобы такого не было, нужно исп функцию copy
b = a[0:3].copy()

# обычные срезы так не делают
a = [1, 2, 3, 5]
b = a[0:3]
b[0] = 10
a[1] = 20
print(a, b)

# так делают обычные массивы
a = [1, 2, 3, 5]
b = a
b[0] = 10
a[1] = 20
print(a, b)
```

```python
d2 = np.array([[1, 2, 3],
               [4, 5, 6]])

# способы обращения по индексу
print(d2[1][0])
print(d2[1, 0])

# взятие столбца
print(d2[:, 0])
print(d2[:, 1])
```

```python
# Пример 2
names = np.array(['Саша', 'Таня', 'Андрей', 'Ваня', 'Аня', 'Андрей'])
print(names[names != 'Андрей'])

mask1 = names != 'Андрей'
mask2 = names != 'Ваня'

print(names[mask1 & mask2]) #логич операции заменили битовыми ~, |, &
```

**np.where(condition, x, y)** - функция для записи условий, condition - условие, если для элем массива условие condition истинно, его значение станет x, иначе y

```python
# Функция where
arr = np.array([1, 2, 30, 40, 5, 6])

res1 = np.where(arr % 2 == 0, arr // 2, arr) # т.е если элемент четный, он делится пополам, иначе остается без изменений

# условия могут быть и составными(операторы ~&| и каждое условие записывается в скобках)
res2 = np.where((arr % 2 == 0) | (arr > 10), arr, 0) #т.е если элем четный или больше 10, он остается без изм, иначе заменяется на 0

print(res1)
print(res2)
```

```python
# условия where могут быть вложенными

# Пример: разбить клиентов на группы: до 25 лет (включительно) - A, с 26 до 40 лет (включительно) - B, старше 40 лет - C
age = np.array([25, 22, 27, 33, 37, 45, 50])
res = np.where(age <= 25, 'A', np.where((age >= 25) & (age <= 40), 'B', 'C'))
#res = np.where(age <= 25, 'A', np.where(age >= 40, 'C', 'B')) # или так
print(res)
```

**np.isin(elements, test_elements)**- функция проверяет находится ли каждый элем массива elements в массиве(листе) test_elements. Возвращается маска

```python
m2d = np.array([[0, 1, 2, 3],
                [1, 3, 0, 2]])

print(np.isin(m2d, [1, 2]))
```

Пусть **mask** - массив булевых значений

Тогда:
**mask.any()** - метод вернет True, если в массиве есть хотя бы один элем True

**mask.all()** - метод вернет True, если в массиве все элем True

```python
bools1 = np.array([True, True, False, True])
bools2 = np.array([False, False, False, False])
bools3 = np.array([True, True, True, True])

print(bools1.any())
print(bools2.any())
print(bools3.any())

print(bools1.all())
print(bools2.all())
print(bools3.all())
```

# Генерация случайных чисел

Функция **np.random.rand(shape)** создает массив заданной размерности shape, заполненный числами от 0 до 1 (единица в диапазон не входит).



Функция **np.random.uniform(a, b, shape)** формирует массив вещественных чисел в заданном диапазоне [a, b) и с заданной размерностью. Границы a, b могут быть заданы любыми числами.

Функция **np.random.randint(a, b, shape)** формирует массив целых чисел в заданном диапазоне [a, b) и с заданной размерностью. Границы a, b должны быть заданы целыми числами.


```python
np.random.rand(5) # 5 случайных чисел из равномерного распределения [0, 1)
```

```python
np.random.rand(4, 3) # # матрица случайных чисел из равномерного распределения [0, 1) размером 4 * 3
```

```python
# создадим массив размерностью 2 x 3 x 2 c числами [-3, 3)
np.random.randint(-3, 3, size=(2, 3, 2))
```

```python
# создадим массив размерностью 2 x 3 x 2 c числами [-3, 3)
np.random.randint(-3, 3, size=(2, 3, 2))
```

```python
np.random.uniform(-5, 5, size=10) # одномерный вектор из 10 вещественных чисел [-5, 5)
```

**np.random.permutation(x**)(где х - целое число) - генерируется массив целых чисел [0, x) с помощью функции np.arange(x) и перемешивается

```python
np.random.permutation(10)
```
