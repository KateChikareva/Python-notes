Функции округления:


1.   **math.floor(x)** - округляет число вниз(floor - "пол")
2.   **math.ceil(x)** - округляет число вверх(ceil - "потолок")
3.   **round(x)** - округляет число до ближайшего целого. Если дробная часть числа равна 0.5, то число округляется до ближайшего четного числа.
4.   **round(x, n)** - округляет число x до n знаков после точки.
5.   **int(x)** - отсекает часть после точки







```python
import math

# округление вниз
f1 = math.floor(4.7)     # f1 = 4
f2 = math.floor(-4.7)    # f2 = -5
print(f1, f2)

# округление вверх
c1 = math.ceil(4.2)     # c1 = 5
c2 = math.ceil(-4.2)    # c2 = -4
print(c1, c2)

# отсечение дробной части
i1 = int(4.7)           # i1 = 4
i2 = int(-4.7)          # i2 = -4
print(i1, i2)

# округление до ближайшего целого
r1 = round(4.49)         # r1 = 4
r2 = round(4.51)         # r2 = 5
r3 = round(4.5)         # r3 = 4
r4 = round(5.5)         # r3 = 6
print(r1, r2, r3, r4)

# округление до n знаков
r1 = round(1 / 3, 2)  # r1 = 0,33
r2 = round(1.6666, 3) # r2 = 1,667
print(r1, r2)
```

# БИТОВЫЕ ОПЕРАЦИИ

В языке Python имеются такие битовые операции:

* Инверсия – ~
* Сдвиг вправо – >>
* Сдвиг влево – <<
* И – &
* ИЛИ – |
* XOR (Исключающее или) – ^

```python
# Инверсия - меняет все 0 на 1 и 1 на 0 в двоичном представлении числа.
# Эквивалентная математическая операция: ~a = -a - 1
print(~3)
# на тип int в Python отводится 4 байт
# 3 = 00000000 00000000 0000000 00000011
# ~3= 11111111 11111111 1111111 11111100 = -4, т.к. первый бит знаковый
```

```python
# Сдвиг - разряды в двоичном представлении числа сдвигаются влево или вправо на указанное кол-во разрядов.
# Сдвиг вправо на n = деление нацело на 2 n раз
# Сдвиг влево на n = умножение на 2 n раз

print(21 >> 3) # 2 (10101 -> 10)
print(21 // 2 // 2 // 2) # 21 >> 3 это то же самое, что поделить нацело на 2 три раза

print(5 << 2) # 20 (101 -> 10100)
print(5 * 2 * 2)  # 5 << 2 это то же самое что умножить на 2 два раза
```

```python
# Битовая конъюнкция (&) и дизъюнкция (|) работают так же как и логическое И и логическое ИЛИ, только в мире двоичных чисел.
print(5 & 3) # 1
#  101
# &
#  011
#  ---
#  001

print(5 | 3) # 7
#  101
# |
#  011
#  ---
#  111

# Исключающее или (^). Если биты разные, тогда при XOR-е будет 1, иначе 0.
print(5 ^ 2) # 7
#  101
# ^
#  010
#  ---
#  111
```

Многострочная f-строка — это способ записи длинной форматированной строки (f-string) на нескольких строках кода для лучшей читаемости. При этом Python автоматически склеит строки, если они записаны через круглые скобки и каждая из них — это строка (в том числе f-строка).

Python воспринимает всё внутри скобок как одну длинную строку, автоматически склеивая части, потому что каждая из них — f-строка. Это удобно для длинных URL или SQL-запросов, чтобы не писать всё в одну неудобочитаемую строку.

```python
city = "Moscow"
units = "metric"
lang = "ru"
api_key = "123fakeapikey456"

url = (
    f"https://api.weather-example.com/data?"  # Каждая часть должна быть f-строкой (начинаться с f"..."), чтобы выражения в фигурных скобках подставлялись корректно
    f"city={city}"
    f"&units={units}"
    f"&lang={lang}"
    f"&key={api_key}"
)  # Склеивание работает только при использовании круглых скобок без знаков +.

print(url)
```

**СРЕЗЫ**

**Срез** (slice) — извлечение из данной строки одного символа или некоторого фрагмента(подстроки)
В строке проводится положительная и отрицательная индексация символов.(положительная - слева направо, начиная с 0, отрицательная - справа налево, начиная с -1)

```python
#взятие последнего символа в строке
line = "Hello"
last_sym1 = line[len(line) - 1] #взятие через положительный индекс
last_sym2 = line[-1] #взятие через отрицательный индекс
print(last_sym1, last_sym2)

```

Срез с двумя параметрами: s[a:b] - срез символов от a до b, не включая b; т.е. берутся (b - a) символов

```python
s = "Hello"
#два положительных индекса
print(s[0:3]) # Hel

#два отрицательных индекса
print(s[-5:-2])  # Hel

#смешенные индексы
print(s[0:-1]) #Hell #от первого до последнего символа, не включаю его
print(s[-4:5]) #ello

print(s[0:-1:-1])
```

# ЦИКЛ FOR

Цикл for в языке программирования Python предназначен для перебора элементов в определенной структуре данных. Множество значений может быть задано списком, кортежем, строкой или диапазоном.

```python
#цикл for перебирает элементы списка, в списке значений могут быть выражения различных типов

#При первых трех итерациях цикла переменная i будет принимать значение типа int, при последующих трех — типа str.
for i in 1, 2, 3, 'one', 'two', 'three':
    print(i)
```

```python
#перебирает символы в строке
s = 'abcdefgh'
for c in s:
  print(c)

#перебирает числа в диапазоне от 0 до len(s), не включая len(s)
for i in range(len(s)):
  print(i, s[i])
```

Функция **range()** возвращает неизменяемую последовательность чисел между заданным начальным целым числом и конечным целым числом. \
**range(start, stop, step)**, т.е последовательность чисел от start до stop с шагом step \
По умолчанию start = 0, step = 1 -> range(6) [0, 1, 2, 3, 4, 5] \


```python
#вывод всех нечетных чисел от 1 до 20
for i in range(1, 21, 2):
  print(i, end=' ')

print()

#вывод всех чисел от 1 до 20 в обратном порядке
for i in range(20, 0, -1):
  print(i, end=' ')
```

Python range() работает только с целыми числами. Для работы с типом float можно использовать функцию arange(start, stop, step) из библиотеки NumPy.

```python
import numpy
for i in numpy.arange(0, 21, 2.5):
  print(i)
```

# BREAK, CONTINUE

Оператор break досрочно прерывает цикл.

```python
#при нахождении пробела цикл перебора останавливается
for i in 'hello world':
  if i == ' ':
    break
  print(i, end = '')
```

Выражение continue дает возможность пропустить часть цикла, где активируется внешнее условие, но при этом выполнить остальную часть цикла. При этом прерывается текущая итерация цикла, но программа возвращается к началу цикла.

```python
#т.е если условие выполняется, то команды, следующие после него, не выполняются, начинается новая итерация цикла
for i in range(6):
  if i == 3:
    continue
  print(i, end = ' ')
```

# ЦИКЛ WHILE и WHILE/ELSE

После тела цикла можно написать слово else. Тогда, стоящий после else блок операций, выполнится один раз после цикла, т.е когда проверяемое условие станет неверно.


В целом, конструкция while-else, будет работать точно также, как и просто while(т.е. инструкции также выполнятся после окончания цикла). См пример 1 и 2.

```python
#пример 1
i = 1
while i <= 5:
    print(i)
    i += 1
else:
    print('Цикл окончен, i =', i)

#пример 2
i = 1
while i <= 5:
    print(i)
    i += 1
print('Цикл окончен, i =', i)
```

Смысл появляется только вместе с инструкцией break. Если во время выполнения Питон встречает инструкцию break внутри цикла, то он сразу же прекращает выполнение этого цикла и выходит из него. При этом ветка else исполняться не будет.

```python
#пример 3: после генерации нуля выполнение цикла продолжается, инструкция else выполняется
import random
i = 0
while i < 10:
  number = random.randint(0, 3)
  print("Число №" + str(i + 1) + ' = ' + str(number))
  if (number == 0):
    print("Найден 0")
  i += 1
else:
  print("Выполнение цикла закончено")
```

```python
#пример 4: после генерации нуля выполнение цикла останавливается, инструкция else не выполняется
import random
i = 0
while i < 10:
  number = random.randint(0, 3)
  print("Число №" + str(i + 1) + ' = ' + str(number))
  if (number == 0):
    print("Найден 0")
    break
  i += 1
else:
  print("Выполнение цикла закончено")
```

Функции **min()** и **max()** могут работать и со списками. Если этим функциям передается несколько списков, то целиком возвращается один из переданных списков. При этом сравнение происходит поэлементно: сначала сравниваются первые элементы списков. Если они не равны, то функция min() вернет тот список, первый элемент которого меньше, max() – наоборот. Если первые элементы равны, то будут сравниваться вторые и т. д.

```python
list1 = [1, 7, 12, 0]
list2 = [1, 7, 90]
list3 = [1, 10]

print(min(list1, list2, list3)) #[1, 7, 12, 0]
print(max(list1, list2, list3)) #[1, 10]
```

СОЗДАНИЕ СПИСКА

```python
#Первый способ:
#Cначала создадим список из n элементов(для начала просто из n нулей), затем сделаем каждый элемент списком из m элементов
n = 3
m = 4
a = [0] * n
for i in range(n):
    a[i] = [0] * m

#Второй способ: создать пустой список, потом n раз добавить в него новый элемент: одномерный список из m элементов
b = []
for i in range(n):
  b.append([0] * m)


#Третий способ: списочное выражение. Создать список из n элементов, каждый из которых будет списком, состоящих из m нулей:
c = [[0] * m for i in range(n)]

print('a =', a)
print('b =', b)
print('c =', c)
```

СЧИТЫВАНИЕ СПИСКА ИЗ ИНПУТА


Если элементы списка вводятся через клавиатуру (каждая строка на отдельной строке, всего n строк, числа в строке разделяются пробелами), для ввода списка можно использовать следующий код. \
*Пример ввода: \
1 2 3 \
4 5 6 \
7 8 9*

```python
#Первый способ
n = int(input('Введите количество строк матрицы: '))
a = []
for _ in range(n):
    row = [int(i) for i in input().split()] #считываем строку, создаем список из ее элементов, приведенных к int
    a.append(row) #добавляем подготовленный список в матрицу
print('a = ', a)
```

```python
#Второй способ, тоже самое, но через списочное выражение
n = int(input('Введите количество строк матрицы: '))
a = []
for i in range(n):
    a.append([int(s) for s in input().split()])
```

ПЕРЕБОР ЭЛЕМЕНТОВ И ВЫВОД СПИСКА НА ЭКРАН

```python
a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
#Первый способ: вложенные циклы с перебором индексов
for i in range(len(a)):
    for j in range(len(a[i])):
        print(a[i][j], end=' ')
    print()

#Второй способ: вложенные циклы с перебором самих элементов
for row in a:
    for elem in row:
        print(elem, end=' ')
    print()

#Третий способ
for i in range(len(a)):
	print(*[a[i]])
```

ВЫРАВНИВАНИЕ ВЫВОДА МАТРИЦЫ С ПОМОЩЬЮ СТРОКОВЫХ МЕТОДОВ \
Если элементы матрицы имеют разное количество разрядов и результат вывода получается смазанным. Для решения проблемы удобно использовать строковые методы ljust() и rjust().


```python
rows, cols = 3, 4                # rows - количество строк, cols - количество столбцов

matrix  = [[277, -930, 11, 0],
           [9, 43, 6, 87],
           [4456, 8, 290, 7]]

for r in range(rows):
    for c in range(cols):
        print(str(matrix[r][c]).ljust(6), end='')
    print()

```

КОПИРОВАНИЕ МАТРИЦ \
\
Поверхностная копия, создаваемая методом copy(),  для вложенных списков не подходит. Для них нужна глубокая копия. \
**Поверхностная копия** создает новый составной объект, и затем (по мере возможности) вставляет в него ссылки на объекты, находящиеся в оригинале. \
**Глубокая копия** создает новый составной объект, и затем рекурсивно вставляет в него копии объектов, находящихся в оригинале. \


```python
import copy
a = [[1, 2, 3], [10, 20, 30]]
surface = a.copy() #поверхностная копия
deep = copy.deepcopy(a)

a[0].append(4)
surface[1].append(40)
deep[0].append(5)

#a[i] и surface[i] ссылаются на один и тот же объект
print(a)
print(surface)

#deep это другой объект
print(deep)

```

# КОРТЕЖИ

**Кортеж(tuple)**- неизменяемый аналог списка \

---

Кортежи поддерживают те же операции, что и списки, за исключением изменяющих содержимое. \
Кортежи поддерживают:
* доступ к элементу по индексу (только для получения значений элементов);
* методы, в частности index(), count();
* встроенные функции, в частности len(), sum(), min() и max();
* срезы;
* оператор принадлежности in;
* операторы конкатенации (+) и повторения (*).


```python
empty_tuple = ()                                      # пустой кортеж
point = (1.5, 6.0)                                    # кортеж из двух чисел
names = ('Timur', 'Ruslan', 'Roman')                  # кортеж из трех строк
nested_tuple = (('one', 'two'), ['three', 'four'])    # кортеж из кортежа и списка

#мы можем создать кортеж на основании строки.
text = 'hello python'
str_tuple = tuple(text)
print(str_tuple)
```

```python
#Для создания кортежа с единственным элементом после значения элемента ставят замыкающую запятую:
my_tuple = (1,)
print(type(my_tuple))
```

*Примечание:* Тот факт, что кортеж является неизменяемым вовсе не означает, что мы не можем поменять содержимое списка в кортеже. \
При этом важно понимать: меняется список, а не кортеж. Списки являются ссылочными типами данных, поэтому в кортеже хранится ссылка на список, которая не меняется при изменении самого списка.

```python
my_tuple = (1, 'python', [1, 2, 3])
print(my_tuple)
my_tuple[2][0] = 100
my_tuple[2].append(17)
print(my_tuple)
```

# МНОЖЕСТВА

**Множества** - структура данных, храняющая в себе только уникальные элементы, неупорядоченные между собой. \
\
Множества удобно использовать для выполнения математических операций: объединение, пересечение, симметрическая разность и т. д.

Элементами множества может быть любой неизменяемый тип данных: числа, строки, кортежи. Изменяемые типы данных не могут быть элементами множества. \
\
Создание множества:




```python
a = {1, 2, 3} #Множетсво из чисел: 1, 2, 3
B = {'Привет', 4, 5} # Множество с разными типами данных

S = set() # это объявление пустого множетсва, но не s = {} тк это объявление пустого словаря

# Для создания множеств может использоваться функция set(),аргументом функции set() может быть список или строка
my_set = set([1, 2, 3, 2])
print(my_set) # Вывод: {1, 2, 3}, тк внутри множества не может быть одинаковых элементов
my_set = set('hello')
print(my_set) #Вывод: {'e', 'l', 'h', 'o'}

```

Множество не поддерживает индексацию и срезы. Вызов my_set[0] приведет к ошибке \
\
Также порядок в котором элементы записываются в множество, не совпадает с тем порядком, в котором эти элементы в множестве хранятся.

```python
#Тк множества не поддерживают индексацию, перебрать все элементы множетсва (в неопределенном порядке) можно при помощи цикла for:
primes = {2, 3, 5, 7, 11}
for num in primes:
    print(num)
```

МЕТОДЫ МНОЖЕСТВ

**s.add(x)** - добавление элемента x в множество s \
**s.update(a)** - добавление в множество s нескольких элементов, аргумент a может быть кортежем, списоком, строкой(тогда в s будут добавляться символы строки) или другим множеством \
**s.remove(x)** - удаляет элемент х из множества s, если элемент x в множестве отсутсвовал, метод вызовет исключение KeyError \
**s.discard(x)** - удаляет элемент х из множества s, если элемент x в множестве отсутсвовал, список останется без изменений \
**s.clear()** - очищает множество (удаляет все его элементы) \
**sorted(s)** - сортировка множетсва, метод возвращает список \
**x in s** - проверка на то, присутсвует ли элемент x в множестве s \
**len(s)** - количество элементов множества \


```python
#Варианты заполнения списка
#1) Создать пустой список и добавлять по одному
s = set()
for i in range(10):
  s.add(i)
print(s)

#2) Создать пустой список и добавлять все сразу
s = set()
s.update([i for i in range(10,20)])
print(s)

#3) Использовать функцию set(), аргументом которой может быть строка или список
result = set('a b c d e'.split())
print(result)

#4) В {} можно передать список
inp = '10 20 39 40 5'
guess = {int(x) for x in inp.split()}
print(guess)
```

```python
#Сортировка
import random
my_set = set(random.randint(1, 100) for i in range(10)) #заполнение множетсва 10 случ числами от 1 до 100
print(my_set) #печать множетсва
print(sorted(my_set)) #сортировка множетсва, метод возвращает список
print(*sorted(my_set)) #print(*) печатает элементы списка в строку через пробел
```

Множества можно сравнивать между собой. Равные множества имеют одинаковую длину и содержат равные элементы. Для сравнения множеств используются операторы == и !=. Порядок элементов не учитывается.

```python
myset1 = {1, 2, 3, 3, 3, 3}
myset2 = {2, 1, 3}
myset3 = {1, 2, 3, 4}

print(myset1 == myset2)
print(myset1 == myset3)
print(myset1 != myset3)
```

ДОСТУП К ЭЛЕМЕНТАМ: \

**d[k]** - доступ к элементу словаря d по ключу k, осуществляется также как и доступ к элементу списка, только в качестве индекса указывается ключ. Обращение к элементу по несуществующему ключу выдаст ошибку KeyError: k


**d.get(k, inf)** - такой же метод для доступа к элементу словаря d по ключу k. При обращении к элементу по несуществующему ключу метод вернет параметр inf, если параметр inf не указан, вернется default (по умолчанию None)


**d.setdefault(k, default)** - метод позволяет получить значение из словаря по заданному ключу k, автоматически добавляя в словарь пару ключ-значение(k: default), если он отсутствует. При этом если значение default не передано в метод, то вставится значение None

```python
#словарь квадратов чисел
d = {1: 1, 2: 4, 3: 9}

print(d[1]) #обращение к элементу по ключу
print(d.get(1)) #такой же доступ к элементу по ключу, только через метод get

#обращение к элементу по несуществующему ключу
#print(d[0]) #выдаст ошибку KeyError: 0
print(d.get(0)) #возвращает default (по умолчанию None)
print(d.get(0, 'Элемент не найден'))


#расширение словаря, новый ключ добавит в словарь новую пару
d[4] = 4 ** 2
print(d) #d == {1: 2, 2: 4, 3: 9, 4: 16}

#перезапись элемента, по имеющемуся ключу записывается новое значение
d[1] = 1
print(d) #d == {1: 1, 2: 4, 3: 9, 4: 16}

print(d.setdefault(5, 25)) #метод вернул 25, значение для ключа 5, но так как такой пары в словаре не было, она добавилась
print(d)
```

```python
#
d = {i: i ** 2 for i in range(4, 11)}

#1 способ
for key in d:
    print('Число:', key, 'Квадрат числа:', d[key])

#2 способ
print('Числа:')
for key in d.keys(): # итерируем по списку [4, 5, 6, 7, 8, 9, 10]
  print(key, end = ' ')
print()

print('Квадраты:')
for v in d.values(): # итерируем по списку [16, 25, 36, 49, 64, 81, 100]
  print(v, end = ' ')
print()

for key, value in d.items():
   print( key,'-' ,value)
```

**Значением словаря может быть другой словарь, т.е. словари могут быть вложенными.** \
*Задача: "Продажи"* \
Дана база данных о продажах некоторого интернет-магазина. Каждая строка входного файла представляет собой запись вида Покупатель товар количество, где Покупатель — имя покупателя (строка без пробелов), товар — название товара (строка без пробелов), количество — количество приобретенных единиц товара.

Создайте список всех покупателей, а для каждого покупателя подсчитайте количество приобретенных им единиц каждого вида товаров. Список покупателей, а также список товаров для каждого покупателя нужно выводить в лексикографическом порядке.

*Пример входных данных:* \
Стас Ручка 1 \
Филипп Ручка 1 \
Артем Перо 3 \
Стас Линейка 4 \
Стас Тетрадь 7 \
Стас Ручка 29 \
Филипп Циркуль 1 \

*Выходные данные:* \
Артем: \
Перо 3 \
Стас: \
Линейка 4 \
Ручка 30 \
Тетрадь 7 \
Филипп: \
Ручка 1 \
Циркуль 1 \

```python
order = {}
while True: #пока данные вводятся
    try:
        name, product, count = input().split()
        count = int(count)
        if name not in order: #если человека нет в списке заказов, то есть первый заказ
          order[name] = {product: count} #в словарь добавляется новый элемент, ключ - имя заказчика, значение - другой словарь {товар: количество}
        else: #если человек есть в списке
          if product not in order[name]: #если человек покупает товар впервые
            order[name][product] = count
          else:
            order[name][product] += count #если человек уже заказывал данный товар
    except:
      break

for i, j in sorted(order.items()): #сортировка по ключу, т.е сортировка фамилий в алфавитном порядке
    print(i,':', sep = '')
    for j in sorted(order[i]): #сортировка списка покупок человека в алфавитном порядке
        print(j, order[i][j]) #выводится: покупка, количество

```

#ГЕНЕРАЦИЯ СЛУЧАЙНЫХ ЧИСЕЛ

**Случайные числа**– последовательность чисел, в которой невозможно предсказать следующее число, зная все предыдущие.

Ставить сложные электронные приборы на каждый компьютер для генерации истинно случайных чисел дорого, поэтому математики и программисты создали алгоритмы получения псевдослучайных чисел.

**Псевдослучайное число** - определение, служащее именно для уточнения природы числа, т.е, что оно просто высчитано по формуле из какого-то предсказуемо меняющегося исходного числа, доступного программе (например, часы системы). Зная первое число и формулу, можно определить любое следующее

Разница между генератором настоящих случайных чисел (ГНСЧ) и генератором псевдослучайных чисел (ГПСЧ) в том, что ГНСЧ использует непредсказуемые физические средства для генерации чисел (например сайт random.org использует шумы атмосферы), а ГПСЧ использует математические алгоритмы (полностью производимые компьютером).

В Python для генерации псевдослучайных чисел используется один из самых совершенных алгоритмов генерации псевдослучайных чисел – *вихрь Мерсенна*. Функции для генерации хранятся в модуле **random** в стандартной библиотеке.

МЕТОДЫ МОДУЛЯ RANDOM

**random.random()** — возвращает псевдослучайное число с плавающей точкой в диапазоне [0.0;1.0)

**random.randint(start, end)** — возвращает псевдослучайное целое число в диапазоне [start, end]; start и end должны быть целыми

**random.uniform(start, end)** — возвращает псевдослучайное вещественное число в диапазоне [start, end]; start и end любые

**random.randrange(start, end, step)** - функция генерирует последовательность чисел в диапазоне [start, end) c шагом step и возвращает из нее случайный элемент. По умолчанию start = 0

**random.seed(a)** - функция фиксирует начальное значение a, используемое для последующей генерации, таким образом позволяет генерировать одни и те же случайные числа при новых запусках. Если начальное значение не было задано, то, по умолчанию, начальным значением генератора является системное время (текущая дата и время)

**random.choice(s)** — возвращает случайный элемент из последовательности s(строки, списка, кортежа)

**random.shuffle(my_list)** - метод перемешимает список my_list случайным образом

**random.sample(sequence, k)** - метод выбирает из последовательности sequence(может быть списком, строкой, кортежем или множеством) k случайных элементов и возвращает их списком.



```python
import random
#принцип работы функции seed
random.seed(20)
print(random.random()) #это всегда число 0.9056396761745207

print(random.random()) #любое случайное число

random.seed(20)
print(random.random()) #снова 0.9056396761745207
```

# МОРЖОВЫЙ И ТЕРНАРНЫЙ ОПЕРАТОРЫ. СПИСОЧНЫЕ ВЫРАЖЕНИЯ

Тернарный оператор - конструкция if else, записанная в одну строку

Моржовый оператор- оператор, позволяющий присвоить значение переменной в выражении(месте, где эта переменная используется)

Генератор списка, множетсва или словаря - цикл, записанный в одну строку


```python
x = int(input())
if x % 2 == 0:
  print('число четное')
else:
  print('число нечетное')

#с использовнаием тернарного оператора выше описанный код можно записать в одну строку
print('число четное' if x % 2 == 0 else 'число нечетное')
```

```python
#использование моржового оператора в той же задаче
print('число четное' if (x:= int(input())) % 2 == 0 else 'число нечетное')
```

Python поддерживает четыре вида генераторов:
* генераторы списков (list comprehension) \
new_list = [выражение for элемент in последовательность if условие]

* генераторы множеств (set comprehension) \
new_set = {выражение for элемент in последовательность if условие}

* генераторы словарей (dict comprehension) \
new_dict = {ключ:значение for (ключ,значение) in dict.items() if условие}

* генераторные выражения (generator expressions) \
new_gen = (выражение for элемент in последовательность if условие)

```python
#Пример: из массива рандомных чисел выделим в новый массив четные элементы
import random
random_numbers = [random.randint(0, 100) for _ in range(10)]
print(random_numbers)

# если в выражении есть только if, то он записывается в конце после цикла
even_numbers = [number for number in random_numbers if number % 2 == 0]  # массив только четных чисел
print(even_numbers)

# если в выражении есть if/else, то это тернарный оператор и он записывается перед циклом
even_numbers2 = [number if number % 2 == 0 else 0 for number in random_numbers]
print(even_numbers2)
```
