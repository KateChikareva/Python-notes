ПАРАМЕТРЫ С ПРОИЗВОЛЬНЫМ КОЛИЧЕСТВОМ АРГУМЕНТОВ УКАЗЫВАЮТСЯ КАК:\
***args** - произвольное число позиционных аргументов. При вызове функции, на место этого параметра передается список аргументов, заключенных в кортеж. Перед *args может быть ноль или более позиционных аргументов. После *args могут идти только именнованные аргументы. (см примеры 1,2 ниже)

 ****kwargs** (от словосочетания keyword arguments) - позволяет вам передавать произвольное число именованных аргументов в функцию. При вызове функции, на его место передается список именованных аргументов заключенных в словарь. Параметр kwargs может быть определен совместно с другим формальным параметром args. Параметр kwargs указывается последним в области определения формальных параметров функции (см пример 3 ниже)

 Имена args и kwargs по негласному соглашению принято использовать в документации Python. На самом деле никто не запрещает брать любые другие имена переменных.

```python
#Пример 1

#собственная функция my_func, принимающая один позиционный аргумент num и *args - произвольное число позиционных аргументов, объединенных в кортеж
def my_func(num, *args):
    print(args)
    print(num)

my_func(17, 'Python', 2, 'C#') #приведенный код связывает с переменной num значение 17, а с переменной args значение кортежа ('Python', 2, 'C#')
my_func(17) #код связывает с переменной num значение 17, а с переменной args значение пустого кортежа ()
```

```python
#Пример 2

#собственная функция my_sum, принимающая произвольное количество позиционных аргументов и возвращающая их сумму
def my_sum(*args):
	res = 0
	for element in args:
		res+= element
	return res


print(sum(1, 2, 3, 4, 5)) #произойдет ошибка, тк встроенная функция sum принимает в качестве аргумента коллекцию - список, сет и тд
print(my_sum(1, 2, 3, 4)) #собственная функция принимает переменное количество аргументов, в данном случае 4

print(sum([1, 2, 3, 4]) ) #правильное использование функции sum
print(my_sum(*[1, 2, 3, 4])) #собственная функция может принимать коллекции, только их надо распаковать

print(my_sum(1, 2, *[3, 4, 5], *(7, 8, 9), 10)) #Более того, часть аргументов можно передавать непосредственно и даже коллекции подставлять не только по одной.
```

```python
#Пример 3
#Собственная функция get_info принимает произвольное количество именнованных аргументов и печатает их
def get_info(**kwargs):
  for key, value in kwargs.items():
      print(key, '=', value)

info = {'name':'Timur', 'age':'28', 'job':'teacher'}
get_info(**info) #1 способ передачи именнованых аргументов в функцию

print()

get_info(name='Kate', age=22, job='teacher', education='programming') #2 способ передачи именнованых аргументов в функцию
```

Сформулируем общее правило, касающееся видимости переменных:

1. Внутри функции видны все переменные этой функции (локальные переменные и аргументы функции).

3. Снаружи не видны никакие переменные, которые определены внутри функции.

Когда вы внутри функции выполняете присваивание, появляется новая внутренняя (локальная) переменная и, если ее имя совпадает с именем внешней, она перекрывает переменную из внешней (глобальной) области видимости — и внешняя переменная остается неизменной, таким образом, внешние переменные защищены от случайной порчи.

Если вы все же хотите изменить внешнюю переменную, вы должны описать переменную внутри функцию ключевым словом global. Но злоупотреблять этим не рекомендуется, поскольку отдельные части программы начинают зависеть друг от друга, что может привести ко множеству ошибок.

```python
#ситуация, когда внутри функции используется переменная с тем же именем, что и в основной программе, но перед использованием ей присваивается новое значение.
area = 'Красная площадь'


def print_square_area(length, width):
    area = length * width. # локальная переменная с таким же именем перекрыла глобальную
    print('Площадь площади: ', area)


print('Место встречи: ', area)
print_square_area(330, 75)
print('Повторяю, место встречи: ', area)
```

```python
# Пример пользы от глобальных переменных

# Функция принимает список чисел и возвращает количество значений, больших предыдущей нормы.
# Предыдущая норма – среднее значения элементов списка, переданного при предыдущем вызове функции. При первом вызове средним считать 0.
k = 0


def changing_the_norm(nums):
    global k
    res = len(list(filter(lambda num: num > k, nums)))
    k = sum(nums) / len(nums)
    return res


print(changing_the_norm([150, 160, 164, 147]))
print(changing_the_norm([170, 162, 156, 153, 159]))
print(changing_the_norm([158, 168, 172]))
```

Поскольку функции тоже объекты, работать с ними можно и как с объектами: записывать их в переменные, передавать в качестве аргументов другим функциям, возвращать из функций и т.д.

Возможность записать функцию в переменную позволяет гибко управлять тем, какую функциональность мы хотим использовать. В одну и ту же переменную можно записать разные варианты поведения и менять их при необходимости. При этом не только не нужно будет менять код по всей программе, но и не придется даже изменять код функций.

Поскольку функции являются обычными объектами, то мы можем хранить их в структурах данных точно так же, как и другие объекты. Например, мы можем создать список или словарь, содержащий функции.



```python
#Пример1

#среднее арифметич
def avg(nums):
    return sum(nums)/len(nums)

funcs = [len, sum, min, avg]

primes = [2, 3, 5, 7, 11]

#для списка primes будут поочередно выводиться длина, сумма, минимум и среднее значения
for func in funcs:
    print(func(primes))
```

```python
#Пример 2
#Задача: На вход программы подается название команды(=функции), которую нужно выполнить. Доступные команды: start, stop, pause

def start():
    print('Выполнена команда start')


def stop():
    print('Выполнена команда stop')

def pause():
    print('Выполнена команда pause')


command = input()        # считываем название команды

#Такую логику легко можно написать при помощи условного оператора if .. elif.
if command == 'start':
    start()
elif command == 'stop':
    stop()
elif command == 'pause':
    pause()

#Однако если команд будет много или, если их количество будет увеличиваться, то оператор if получится слишком громоздким.
#В этом случае можно создать словарь, где ключом служит название команды, а значением — соответствующая функция.

commands = {'start': start, 'stop': stop, 'pause': pause}  # словарь соответствия команда → функция
commands[command]()      # вызываем нужную функцию через словарь по ключу
```

**ФУНКЦИИ В КАЧЕСТВЕ АРГУМЕНТОВ ДРУГИМ ФУНКЦИЯМ**

Поскольку функции являются объектами, можно передавать их в качестве аргументов другим функциям.

```python
#Пример:
def talk(command, name):
    fraza = command(name)
    print(fraza)

def hello(name):
    return f'Привет {name}.'

def goodbye(name):
    return f'Пока {name}.'

#Функции hello и goodbye передается функции talk() без скобок. Это означает, что передается только ссылка на функцию, а функция при этом не выполняется.
#Функция talk(), записывается со скобками, поэтому она будет вызываться как обычно.
talk(hello, 'Иван')
talk(goodbye, 'Петр')
```

# **ФУНКЦИИ ВЫСШЕГО ПОРЯДКА**

Функции высшего порядка - функции, которые принимают или/и возвращают другие функции.

Встроенные функции **max, min, sort, sorted** являются функциями высшего порядка, тк имеют необязательный параметр** key **- функцию, определяющую условия сравнения элементов. Другими словами, значение key должно быть функцией, принимающей один аргумент и возвращающей на его основе ключ для сравнения. \
 Функция, определяющая условия сравнения элементов, называется компаратор (compare – сравнивать).

```python
#Пример 1
numbers = [10, -7, 8, -100, -50, 32, 87, 117, -210]
print(sorted(numbers, key=abs))     #  указываем функцию abs в качестве компаратора, т.е. Исходный массив будет отсортирован по возрастанию модуля чисел
print(max(numbers, key=abs)) #поиск максимального по модулю элемента
```

# **АНОНИМНЫЕ ФУНКЦИИ**

Помимо стандартного определения функции, состоящего из ее заголовка с ключевым словом def и телом функции  Python можно создавать короткие однострочные функции с использованием оператора lambda.

**Анонимные функции(=лямбда функции)** – функции с телом, но без имени

Общий формат определения анонимной функции: \
*lambda список_параметров: выражение*

```python
#Пример 1: Следующие два определения функций эквивалентны

def standard_function(x):            #  стандартное объявление функции
    return x * 2

lambda_function = lambda x: x * 2      #  объявление анонимной функции

print(standard_function(5))
print(lambda_function(5))
```

```python
#Пример 2
f1 = lambda: 10 + 20               # функция без параметров
f2 = lambda х, у: х + у            # функция с двумя параметрами
f3 = lambda х, у, z: х + у + z     # функция с тремя параметрами

print(f1())
print(f2(5, 10))
print(f3(5, 10, 30))
```

В теле анонимной функции не получится выполнить несколько действий и не получится использовать многострочные конструкции вроде циклов for и while. Однако можно использовать тернарный условный оператор.

Общий вид тернарного условного оператора в теле анонимной функции выглядит так: \
*значение1 if условие else значение2*


```python
#Пример:
numbers = [-2, 0, 1, 2, 17, 4, 5, 6]
result = list(map(lambda x: 'четное' if x % 2 == 0 else 'нечетное', numbers))
print(result)
```

Интересная особенность анонимных функций (лямбда-функций) – они являются выражениями. После определения лямбда-функции ее можно сразу же вызвать.

```python
print((lambda х, у: х + у)(5, 10))     # 5 + 10
print(1 + (lambda x: x*5)(10) + 2)     # 1 + 50 + 2
```

Применение анонимных функций, как правило, оправдано в следующих ситуациях:
* однократное использование функции
* передача функций в качестве аргументов другим функциям
* возвращение функции в качестве результата другой функции

Анонимные функции очень часто используются вместе со встроенными функциями map(), filter(), reduce(), sorted(), max(), min() и т.д.

```python
#Пример 1
numbers = [10, -7, 8, -100, -50, 32, 87, 117, -210]

print(max(numbers, key=abs)) #  поиск элемента, модуль которого максимален
print(min(numbers, key=lambda x: x**2)) #  поиск элемента, квадрат которого минимален
print(sorted(numbers, key=lambda x: -x)) #  сортировка по убыванию
```

```python
#Пример 2, аналогичный пример с функциями компараторами есть выше

#список кортежей; координат точек на двухмерной плоскости
points = [(1, -1), (2, 3), (-10, 15), (10, 9), (7, 18), (1, 5), (2, -4)]

print(sorted(points, key=lambda pair: pair[1])) #сортируем по второму значению кортежа, в случае их совпадения, исходный порядок не меняется
print(sorted(points, key=lambda pair: pair[0] + pair[1])) #сортируем по сумме кортежа,  в случае  совпадения сумм, исходный порядок кортежей не меняется
print(sorted(points, key=lambda pair: (pair[0] + pair[1], pair[0]))) #сортируем по сумме кортежа,  в случае  совпадения сумм, по возрастанию первых элементов кортежей
```

**map(func, collection)** - функция возвращает итератор по всем элементам коллекции collection, преобразованным c помощью функции трансформации func; т.е. map к каждому элементу collection применяет функцию func (см пример 1,2)

**all(iterable)** - функция вернет True если, все элементы переданной ей последовательности iterable  истинны, , и False в противном случае.

**any(iterable)** - функция вернет True, если хотя бы  один элемент переданной ей последовательности iterable  является истинным , и False в противном случае.

в Python все следующие значения приводятся к значению False:
* константы None и False;
* нули всех числовых типов данных: 0, 0.0, 0j, Decimal(0), Fraction(0, 1);

Функции *all() и any()* могут быть полезны в комбинации с функцией map(), которая может преобразовывать элементы последовательности (итерируемого объекта) к значению True/False в соответствии с неким условием.

# **АТРИБУТ ФУНКЦИИ \_\_dict\_\_**

У объектов функций есть дополнительный атрибут **\_\_dict\_\_**, являющийся словарем который используется для динамического наделения функций дополнительным функционалом. Доступ к словарю атрибутов функции можно получить как из тела функции, так и из вне.

Устанавливать и получать значения из данного атрибута можно используя два синтаксиса:
* в стиле словаря: **func.\_\_dict\_\_['attr'] = value**
* через точечную нотацию: **func.attr = value**

```python
#Пример 1
def greet():
    greet.age = 17

print(greet.__dict__) #  словарь атрибутов функции greet() пустой

greet.value = 777
greet.numbers = [1, 2, 3]
greet.name = 'Timur'

print(greet.__dict__)  #к словарю атрибутов функции greet()  добавилось три пары значений

greet()

print(greet.__dict__) #из-за вызова функции greet к словарю добавилась пара age = 17
```

Словарь атрибутов может быть использован для кэширования уже вычисленных значений функции.

```python
#Пример 2:

#Рекурсивное вычисление чисел фибоначи без кэширования
def fib1(num):
    if num < 2:
        return num
    return fib1(num - 1) + fib1(num - 2)

#Рекурсивное вычисление чисел фибоначи с кэшированием ранее вычисленных значений в словаре атрибутов
def fib2(num):
    if num < 2:
        return num
    if num not in fib2.__dict__: #если значение функции не было вычисленно и закешировано ранее
        fib2.num = fib2(num - 1) + fib2(num - 2) #оно вычисляется и кэшируется
    return fib2.num #затем возвращается

print(fib1(5))
print(fib2(5))
```

# **ВЛОЖЕННЫЕ ФУНКЦИИ**

Функции могут быть вложенными. \

Существует 2 варианта вложенности:
1.   Когда функция вызывает функцию внутри себя
2.   Когда функция возвращает вложенную в нее функцию



```python
#1 пример: функция вызывает функцию внутри себя
def speak(text):
    def whisper(t):                      # объявляем вложенную функцию
        return t.lower() + '...'
    return whisper(text)                 # вызываем вложенную функцию и возвращаем ее результат

print(speak('Hello, World'))
```

```python
#2 пример: когда функция возвращает вложенную в нее функцию
def get_speak_func(volume):
    def whisper(text):
        return text.lower() + '...'
    def yell(text):
        return text.upper() + '!'
    #функция на самом деле не вызывает одну из своих вложенных функций
    #она просто выбирает подходящую функцию на основе аргумента volume, а затем возвращает объект этой функции
    if volume > 0.5:
        return yell
    else:
        return whisper

whisper = get_speak_func(0.3)     # функция whisper()
yell = get_speak_func(0.7)        # функция yell()

print(whisper('Hello'))           # говорим шепотом
print(yell('Hello'))              # кричим
```

# **ЗАМЫКАНИЯ**


Замыкание - это особый вид функции, при котором:
- Замыкание это вложенная функция, т.е находящаяся целиком в теле другой функции
- Оно содержит ссылки на локальные переменные внешней функции
- Замыкание является возвращаемым значением родительской функцией
- Каждый раз при выполнении внешней функции происходит создание нового экземпляра внутренней функции, с новыми ссылками на переменные внешней функции.

```python
#Пример замыкания: greeting_creator - функция для настройки приветсвия
#greeting_word - приветственное слово, (Здравствуйте, Привет, Добрый день и тд)
def greeting_creator(greeting_word):
    def greet(name):
        return f'{greeting_word}, {name}' # вложенная функция greet() может обращаться к аргументу greeting_word своей родительской функции

    return greet

say_hi = greeting_creator('Hi')
say_hello = greeting_creator('Hello')

print(say_hi('Timur'))
print(say_hello('Soslan'))
```

Замыкания могут обращаться ко всем локальными переменным родительской функции, это могут быть переменные, объявленные внутри внешней функции или аргументы внешней функции.

Замыкания могут обращаться к таким переменным, но не менять их!

Если внутренняя функция хочет менять переменные внешней функции, она должна объявить ее nonlocal.

```python
def outer(x):
    y = 10
    def inner(z): #замыкание inner обращается к переменным x, y родительской функции
        t = 20
        return (x + y) - (t + z)
    return inner
outer(10)
```

```python
def outer(x):
    def inner():
        x += 10 #замыкание inner пытается изменить переменную внешней функции, такой код приведет к ошибке
        print(x)
    inner()

outer(10)
```

```python
def outer(x):
    def inner():
        nonlocal x #правильное изменение нелокальной переменной x внутри замыкания
        x += 10
        print(x)
    inner()

outer(10)
```

```python
def outer(x):
    def inner():
        x = 20 # при этом так сделать можно, здесь x это локальная переменная функции inner
        print(x)
    inner()

outer(10)
```

При поиске переменной с указанным именем приоритет (правило разрешения имен) следующий:
1. сначала ищем локальную переменную
2. если не нашли локальную, ищем нелокальную
3. если не нашли нелокальную, ищем глобальную
4. и в самом конце — встроенную в язык

# **ДЕКОРАТОРЫ**

**Декоратор** — это функция, которая позволяет изменить поведение другой функции   без непосредственного изменения её кода.

Декоратор принимает другую функцию в качестве аргумента, расширяет ее поведение, не изменяя ее явно, и возвращает новую функцию.


```python
#Пример

# sample_decorator- декоратор, функция высшего порядка
# в качестве аргумента она принимает функцию func, поведение которой хочет изменить, возвращает modification - функцию с новым поведением
def sample_decorator(func):
    def modification(): # это функция, которая обёртывает передаваемую func и затем изменяет её поведение.
        print('Начало функции')
        func()
        print('Конец функции')
    return modification

def say():
    print('Привет Мир!')

new_say = sample_decorator(say)          # декорируем функцию say

say()                                # вызов обычной функции say
new_say() #вызов декорированной функции say
```

Если мы хотим декорировать функцию, которая принимает аргументы, внутренняя функция декоратора тоже должна принимать эти же аргументы. Для универсальности декоратора внутренняя функция может принимать произвольное число позиционных и именованных аргументов \*args и **kwargs

```python
#Решение проблемы, теперь upper_decorator универсален
#будет работать как для функций, которые вообще не принимают аргументы, так и для функций которые принимают произвольное количество позиционных и именованных аргументов.
def upper_decorator(func):
  def change(*args, **kwargs):
      modified_result = str(func(*args, **kwargs)).upper()
      return modified_result
  return change

def hello(name):
    return f'Hello {name}!'

def hello_full(first_name, second_name):
    return f'Hello {first_name} {second_name}!'

hello = upper_decorator(hello)
hello_full = upper_decorator(hello_full)

print(hello('Timur'))
print(hello_full('Ivan', 'Petrov'))
```

ВОЗВРАТ ЗНАЧЕНИЙ ИЗ ДЕКОРИРУЕМОЙ ФУНКЦИИ

 Нужно быть очень внимательным к декорированию функций, которые возвращают значения, которые затем используются.


```python
def double(func): #декоратор, который должен был возвращать удвоенный код цвета
    def wrapper(*args, **kwargs):
        doubled_result = func(*args, **kwargs) * 2
        #return doubled_result #строка которой не хватает, чтобы декоратор работал правильно
    return wrapper

@double
def get_color_code(color):
    color_codes = {'black': '#000000', 'white': '#FFFFFF'}
    return color_codes[color]

print(get_color_code('white')) #однако данный код выводит "None", тк декоратор поглатил возвращаемое значение: вложенная функция wrapper() явно не возвращает какого либо значения
```

ОБЩИЙ ШАБЛОН ДЕКОРАТОРА

На основе всей информации, описанной выше, наиболее частый шаблон декоратора выглядит следующим образом:



```python
import functools

def decorator(func):
    @functools.wraps(func) #строка, чтобы предотвратить перетирание атрибутов __name__ и __doc__ декорируемой функции
    def wrapper(*args, **kwargs):
        # Что-то выполняется до вызова декорируемой функции
        value = func(*args, **kwargs)
        # декорируется возвращаемое значение функции
        # или что-то выполняется после вызова декорируемой функции
        return value
    return wrapper
```

ДЕКОРАТОРЫ С АРГУМЕНТАМИ


```python
#Пример: 2 декоратора stars и lines, которые выполняют очень похожие действия
import functools

def stars(func): # декоратор который печатает 30 символов '*' перед результатом оборачиваемой функции
    def wrapper(*args, **kwargs):
        print('*' * 30)
        return func(*args, **kwargs)
    return wrapper

def lines(func): # декоратор который печатает 10 символов '-' перед результатом оборачиваемой функции
    def wrapper(*args, **kwargs):
        print('-' * 10)
        return func(*args, **kwargs)
    return wrapper

#функции сложения и вычитания, которую будем декорировать
@stars
def add(a, b):
    return a + b

@lines
def diff(a, b):
    return a - b

print(add(2, 3))
print(diff(7, 3))
```

В этой и других ситуациях куда удобнее иметь один декоратор, который можно настраивать, передавая ему соответствующие аргументы.

 Для того чтобы создать декоратор, принимающий аргументы, необходимо добавить еще один уровень вложенности, то есть создать функцию, которая возвращает нужный декоратор:

```python
#декоратор с аргументами
def print_symbols(symbol, length): #функция, принимающая аргументы symbol-символ для печати, length - количество символов
    def decorator(func):
        def wrapper(*args, **kwargs):
            print(symbol * length)
            return func(*args, **kwargs)
        return wrapper
    return decorator


@print_symbols('*', 30)
def add(a, b):
    return a + b


@print_symbols('=', 40)
def diff(a, b):
    return a - b

print(add(3, 9))
print(diff(100, 1))
```

ПРИМЕРЫ ПОЛЕЗНЫХ ДЕКОРАТОРОВ



Декоратор, который подсчитывает, сколько раз вызывается функция. Для сохранения состояния счетчика будем использовать пользовательский атрибут функции.

```python
import functools

def counter(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        wrapper.num += 1
        print(f'Вызов {func.__name__}: {wrapper.num}')
        val = func(*args, **kwargs)
        return val
    wrapper.num = 0
    return wrapper

@counter
def greet(name):
    return f'Hello {name}!'

print(greet('Timur'))
print(greet('Ruslan'))
print(greet('Arthur'))
print(greet('Gvido'))
```

Декоратор timer, который подсчитывает время выполнения функции. Для более точного подсчета декоратор принимает аргумент iters, который задает количество измерений.



```python
import functools, time

def timer(iters=1):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            total = 0
            for i in range(iters):
                start = time.perf_counter()
                value = func(*args, **kwargs)
                end = time.perf_counter()
                total += end - start
            print(f'Среднее время выполнения {func.__name__}: {round(total/iters, 4)} сек.')
            return value
        return wrapper
    return decorator


@timer(iters=1000)
def test(n):
    return sum([(i/99)**2 for i in range(n)])

@timer(iters=3)
def sleep(n):
    time.sleep(n)

res1 = test(10000)
res2 = sleep(4)

print(f'Результат функции test = {res1}')
print(f'Результат функции sleep = {res2}'
```
